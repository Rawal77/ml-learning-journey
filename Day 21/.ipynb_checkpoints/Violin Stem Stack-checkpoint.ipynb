# Violin Stem Stack

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

### Violin Plot

Violin plots show the distribution of data and are useful for comparing multiple distributions.

They combine box plots and KDE (Kernel Density Estimation) to give insights into data shape.

```python
# Generating a random dataset

a = np.random.randint(5, 70, 20)
plt.violinplot(a, showmedians=True)  # Show median for better insights
plt.title("Violin Plot Example")
plt.ylabel("Values")
plt.show()
```

```python
# Using a real-world dataset (Dinner data) to analyze calorie distribution

df = pd.read_csv('../data_sets/Dinner_data.csv')
df.head()
```

```python
plt.violinplot(df['Calories'], showmedians=True)
plt.title("Violin Plot of Calories in Dinner Data")
plt.ylabel("Calories")
plt.show()
```

```python
plt.violinplot(df['Calories'], showmeans=True)  # Show mean instead of median
plt.title("Violin Plot with Mean Indicated")
plt.ylabel("Calories")
plt.show()
```

### Stemplot

Stemplots are useful for visualizing small datasets and identifying patterns in the distribution.

```python
# Generating random data

a = np.random.randint(5, 70, 20)
print("Dataset:", a)
plt.stem(a)
plt.title("Basic Stemplot Example")
plt.ylabel("Values")
plt.show()
```

```python
# Customizing the stemplot with linefmt, markerfmt, and bottom positioning

a = np.random.randint(5, 70, 20)
print("Dataset:", a)
plt.stem(a, linefmt='--', markerfmt='*', bottom=5, orientation='horizontal')
plt.title("Customized Horizontal Stemplot")
plt.xlabel("Values")
plt.show()
```

```python
# Applying stemplot on a real-world dataset

data = pd.read_csv('../data_sets/Titanic-Dataset.csv')
plt.stem(data['Age'].dropna())  # Dropping NaN values for clean visualization
plt.title("Stemplot of Passenger Ages in Titanic Dataset")
plt.ylabel("Age")
plt.show()
```

### Stack Plot

Stack plots are useful for showing cumulative data over time, such as attendance trends or financial data.

```python
days = [1, 2, 3, 4, 5, 6, 7]
people_attended_1 = np.random.randint(1, 40, 7)
people_attended_2 = np.random.randint(1, 30, 7)
people_attended_3 = np.random.randint(1, 50, 7)
colors = ['red', 'yellow', 'green']

plt.stackplot(days, people_attended_1, people_attended_2, people_attended_3, baseline='sym', colors=colors, labels=['Week1', 'Week2', 'Week3'])
plt.legend()
plt.title("Stack Plot of Attendance Over Weeks")
plt.xlabel("Days")
plt.ylabel("Number of Attendees")
plt.show()
```

```python
# Applying stack plot to a real-world dataset

data = pd.read_csv('../data_sets/Lunch_data.csv')
data = data.tail(5)
data
```

```python
# Assume columns represent attendance counts for different categories

plt.stackplot(data['Unnamed: 0'], data['Breakfast'], data['Lunch'], data['Dinner'], colors=['blue', 'orange', 'purple'], labels=['Breakfast', 'Lunch', 'Dinner'])
plt.title("Stack Plot of Meal Consumption Over Time")
plt.ylabel("Number of Meals")
plt.xlabel("Food Number")
plt.legend()
plt.show()
```

```python
# Additional Example - Comparing nutrient intake over time

plt.stackplot(data['Unnamed: 0'], data['Proteins'], data['Carbohydrates'], data['Fats'], colors=['red', 'yellow', 'green'], labels=['Proteins', 'Carbohydrates', 'Fats'])
plt.title("Stack Plot of Nutrient Intake Over Time")
plt.xlabel("Food Number")
plt.ylabel("Nutrient Intake (g)")
plt.legend()
plt.show()
```
