# Seaborn

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

### **Introduction to Seaborn**

Seaborn is a Python visualization library based on Matplotlib. 

It provides a high-level interface for drawing attractive and informative statistical graphics.

```python
# Load a sample dataset

iris = sns.load_dataset("iris")
tips = sns.load_dataset("tips")
```

```python
# **1. Line Plot** - Visualizing trends over time

plt.figure(figsize=(8,5))
sns.lineplot(x=tips["size"], y=tips["total_bill"], marker='o', ci=None)
plt.title("Line Plot: Total Bill vs Size")
plt.xlabel("Size")
plt.ylabel("Total Bill")
plt.show()
```

```python
# **2. Bar Plot** - Comparing categories

plt.figure(figsize=(8,5))
sns.barplot(x="day", y="total_bill", data=tips, estimator=np.mean, ci=None)
plt.title("Bar Plot: Average Total Bill Per Day")
plt.xlabel("Day")
plt.ylabel("Average Total Bill")
plt.show()
```

```python
# **3. Scatter Plot** - Relationship between two numerical variables

plt.figure(figsize=(8,5))
sns.scatterplot(x="total_bill", y="tip", hue="sex", style="time", data=tips)
plt.title("Scatter Plot: Total Bill vs Tip")
plt.xlabel("Total Bill")
plt.ylabel("Tip")
plt.show()
```

```python
# **4. Box Plot** - Showing distribution and outliers

plt.figure(figsize=(8,5))
sns.boxplot(x="day", y="total_bill", data=tips, hue="sex")
plt.title("Box Plot: Total Bill Distribution by Day and Gender")
plt.xlabel("Day")
plt.ylabel("Total Bill")
plt.show()
```

```python
# **5. Violin Plot** - Combination of KDE and Box Plot

plt.figure(figsize=(8,5))
sns.violinplot(x="day", y="total_bill", data=tips, hue="sex", split=True)
plt.title("Violin Plot: Total Bill Distribution by Day and Gender")
plt.xlabel("Day")
plt.ylabel("Total Bill")
plt.show()
```

```python
# **6. Pair Plot** - Pairwise relationships in a dataset

sns.pairplot(iris, hue="species")
plt.show()
```

```python
# **7. Heatmap** - Correlation matrix visualization

# Select only numeric columns
numeric_df = iris.select_dtypes(include=[np.number])

# Create a heatmap
plt.figure(figsize=(10, 6))
sns.heatmap(numeric_df.corr(), annot=True, cmap='coolwarm', linewidths=0.5)
plt.title("Heatmap of Numerical Data Correlation")
plt.show()
```

```python
# **8. Real-world Example: Sales Data Analysis**

# Creating a mock dataset for visualization
sales_data = pd.DataFrame({
    "Month": ["Jan", "Feb", "Mar", "Apr", "May", "Jun", "Jul", "Aug", "Sep", "Oct", "Nov", "Dec"],
    "Sales_A": np.random.randint(200, 500, 12),
    "Sales_B": np.random.randint(150, 450, 12)
})

plt.figure(figsize=(8,5))
sns.lineplot(x=sales_data["Month"], y=sales_data["Sales_A"], marker='o', label="Product A")
sns.lineplot(x=sales_data["Month"], y=sales_data["Sales_B"], marker='s', label="Product B")
plt.title("Sales Trend Over Months")
plt.xlabel("Month")
plt.ylabel("Sales")
plt.legend()
plt.show()
```
