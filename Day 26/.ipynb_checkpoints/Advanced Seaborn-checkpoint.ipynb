# Advanced Seaborn Visualization

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
# Set Seaborn theme for better aesthetics

sns.set_theme(style="whitegrid")
```

### 1Ô∏è Categorical Data Visualization: Stripplot and Swarmplot

Stripplot displays individual data points with jitter, ideal for small datasets.
    
Swarmplot avoids overlap by adjusting positions, offering a clearer view of distribution.


```python
# Creating a sample dataset: Exam scores by section

np.random.seed(42)
sections = np.repeat(['Section A', 'Section B', 'Section C'], 30)
scores = np.concatenate([np.random.normal(70, 10, 30),
                         np.random.normal(75, 8, 30),
                         np.random.normal(65, 12, 30)])
data = pd.DataFrame({'Section': sections, 'Score': scores})
```

```python
# Stripplot Example

plt.figure(figsize=(8,5))
sns.stripplot(x="Section", y="Score", data=data, jitter=True, color="skyblue")
plt.title("Stripplot: Exam Scores by Section")
plt.xlabel("Section")
plt.ylabel("Score")
plt.show()
```

```python
# Swarmplot Example (better for avoiding overlapping points)

plt.figure(figsize=(8,5))
sns.swarmplot(x="Section", y="Score", data=data, palette="Set2")
plt.title("Swarmplot: Exam Scores by Section")
plt.xlabel("Section")
plt.ylabel("Score")
plt.show()
```

# 2Ô∏è Advanced Relationship Plots: lmplot and residplot

lmplot fits and plots a linear regression model, useful for exploring relationships.

residplot shows the residuals of a linear regression to assess model fit.

```python
# Using the 'tips' dataset for relationship analysis

tips = sns.load_dataset("tips")
```

```python
# lmplot Example: Total bill vs Tip with regression line

sns.lmplot(x="total_bill", y="tip", hue="time", data=tips, aspect=1.5, markers=["o", "s"])
plt.title("lmplot: Total Bill vs Tip with Regression")
plt.xlabel("Total Bill")
plt.ylabel("Tip")
plt.show()
```

```python
# residplot Example: Visualizing residuals

plt.figure(figsize=(8,5))
sns.residplot(x="total_bill", y="tip", data=tips, lowess=True, color="purple")
plt.title("Residplot: Residuals of Total Bill vs Tip")
plt.xlabel("Total Bill")
plt.ylabel("Residuals")
plt.show()
```

# 3 Multi-variable & Grid-Based Plots: FacetGrid

FacetGrid creates a grid of plots based on categorical variables.

It's excellent for comparing multiple subsets of a dataset simultaneously.

```python
# Using the 'iris' dataset to compare species distributions

iris = sns.load_dataset("iris")

# FacetGrid Example: Scatter plots of sepal dimensions for each species

g = sns.FacetGrid(iris, col="species", height=4, aspect=1)
g.map(sns.scatterplot, "sepal_length", "sepal_width", color="green")
g.add_legend()
g.set_axis_labels("Sepal Length", "Sepal Width")
g.fig.suptitle("FacetGrid: Sepal Dimensions by Species", y=1.05)
plt.show()
```
