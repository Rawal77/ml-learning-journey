### Box Plot (Box and Whisker Plot)

A Box Plot is a graphical representation of the distribution of a dataset. It highlights key statistical measures:

Minimum (Lower Fence): The smallest value within the 1.5 IQR range.
25th Percentile (Q1, First Quartile): The median of the lower half of the dataset.
Median (Q2, Second Quartile): The middle value that divides the dataset in half.
75th Percentile (Q3, Third Quartile): The median of the upper half of the dataset.
Maximum (Upper Fence): The largest value within the 1.5 IQR range.
Outliers: Data points beyond the upper and lower fences, represented as individual dots.

### Creating a Simple Box Plot

```python
import matplotlib.pyplot as plt
import numpy as np

# Sample data representing exam scores
scores = [25, 53, 53, 56, 75, 32, 54, 24, 43, 65, 42, 53, 54, 42, 23, 54, 54]

# Creating a Box Plot
plt.boxplot(scores)

# Labels and Title
plt.xlabel("Exam Scores")
plt.ylabel("Values")
plt.title("Box Plot of Exam Scores")

# Display the plot
plt.show()
```

###  Box Plot with Multiple Data Sets

```python
import numpy as np
import matplotlib.pyplot as plt

# Generating random test scores for two different exams
exam_1 = np.random.randint(1, 20, 10)
exam_2 = np.random.randint(5, 25, 10)

# Combining both exams into a single plot
plot_values = [exam_1, exam_2]

# Creating the Box Plot
plt.boxplot(plot_values, labels=['Exam 1', 'Exam 2'])

# Labels and Title
plt.xlabel("Exams")
plt.ylabel("Scores")
plt.title("Comparison of Two Exams using Box Plot")

plt.show()
```

### Box Plot on Real Dataset (Expenses Analysis)

### Histogram (Distribution of Data)

A Histogram is a type of bar graph that represents the distribution of numerical data by dividing it into intervals (bins).

### Creating a Simple Histogram

```python
import matplotlib.pyplot as plt
import numpy as np

# Generating 20 random data points between 0 to 40
data = np.random.randint(0, 40, 20)

# Creating a Histogram with 10 bins
plt.hist(data, bins=10, color='blue', edgecolor='black')

# Labels and Title
plt.xlabel("Value Range")
plt.ylabel("Frequency")
plt.title("Simple Histogram")

plt.show()
```

### Customizing a Histogram

```python
import numpy as np
import matplotlib.pyplot as plt

# Generating random dataset
data = np.random.randint(0, 40, 20)

# Creating a histogram with red color
plt.hist(data, bins=10, color='red', edgecolor='black', alpha=0.7)

# Labels and Title
plt.xlabel("Value Range")
plt.ylabel("Frequency")
plt.title("Customized Histogram")

plt.show()
```

### Histogram on Real Dataset (Titanic Age Distribution)

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load dataset
data = pd.read_csv('../data_sets/Titanic-Dataset.csv')

# Creating a histogram for Age column
plt.hist(data['Age'].dropna(), bins=40, color='green', edgecolor='black')

# Labels and Title
plt.xlabel("Age")
plt.ylabel("Frequency")
plt.title("Age Distribution of Titanic Passengers")

plt.show()
```
