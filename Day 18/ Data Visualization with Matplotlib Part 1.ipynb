#  Data Visualization with Matplotlib (Part 1)

Matplotlib is a powerful data visualization library in Python. It helps in representing data graphically to identify trends, patterns, and insights easily. In real-world applications, businesses use it to analyze sales, marketing data, customer behavior, and financial transactions.

### Bar Chart (Bar Plot)

A bar chart is used to compare different categories using rectangular bars. The height of each bar represents the quantity or frequency of a variable.

```python
import matplotlib.pyplot as plt

# Data: No. of students present in each section
y = [100, 65, 88, 95, 78]  
x = ['A', 'B', 'C', 'D', 'E']  

# Creating a bar chart
plt.bar(x, y)

# Labeling axes and title
plt.xlabel("Class Section")
plt.ylabel("No of Students Present")
plt.title("Students in a Class")

# Display the bar chart
plt.show()
```

### Bar Chart with Real Dataset (Expense Analysis)

```python
import pandas as pd
import matplotlib.pyplot as plt

# Read data from a CSV file
df = pd.read_csv('../data_sets/expense_data.csv')

# Display first few rows
df.head()

# Plot bar chart: x-axis = Payment Mode, y-axis = Amount spent
plt.bar(df['Payment Mode'], df['Amount'])

plt.xlabel("Payment Mode")
plt.ylabel("Amount")


# Display chart
plt.show()
```

###  Line Plot

A line plot is used to visualize trends over time, connecting data points with a line.

```python
# Data: No. of customers in a restaurant
y = [100, 55, 48, 35, 28, 50, 82]  
x = ['Sunday', 'Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday']  

# Creating a line plot
plt.plot(x, y, marker='D')

# Labeling axes and title
plt.xlabel("Week")
plt.ylabel("No of Customers Present")
plt.title("Customers in a Restaurant")

# Display plot
plt.show()
```

### Comparing Two Weeks of Restaurant Data

```python
# Data for two different weeks

y1 = [100, 55, 48, 35, 28, 50, 82]  
y2 = [105, 65, 38, 38, 28, 43, 92]  
x = ['Sunday', 'Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday']  

# Plot first week
plt.plot(x, y1, marker='D', ls='--', color='red', label='Week 1')

# Plot second week
plt.plot(x, y2, marker='*', ls='-', color='blue', label='Week 2', alpha=0.5)

# Labels and title
plt.xlabel("Week")
plt.ylabel("No of Customers Present")
plt.title("Customers in a Restaurant")
plt.legend()

# Display plot
plt.show()
```

###  Scatter Plot

A scatter plot is used to show relationships between two numerical variables. Each point represents an observation.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Generate random data for 50 points
x = np.random.randint(1, 10, 50)  # Random numbers for x-axis
y = np.random.randint(10, 100, 50)  # Random numbers for y-axis

# Create scatter plot
plt.scatter(x, y)

# Display plot
plt.show()
```

### Scatter Plot with Random Colors and Sizes

```python
# Generating random data

x = np.random.randint(1, 10, 50)
y = np.random.randint(10, 100, 50)
color = np.random.randint(1, 80, 50)  # Random colors
size = np.random.randint(10, 100, 50)  # Random sizes

# Create scatter plot with varying colors and sizes
plt.scatter(x, y, marker='*', cmap='viridis', c=color, s=size)

# Add color bar
plt.colorbar()

# Display plot
plt.show()
```

### Scatter Plot with Real Dataset (Employee Data)

```python
# Load employee dataset

data = pd.read_csv('../data_sets/employee.csv')

# Display first few rows
data.head()

# Scatter plot of Age vs. Employee ID
plt.scatter(data["age"], data["employee_id"])

# Display plot
plt.show()

# Scatter plot of Age vs. Employee ID with number of trainings as size
plt.scatter(data["age"], data["employee_id"], s=data['no_of_trainings'])

# Display plot
plt.show()
```
