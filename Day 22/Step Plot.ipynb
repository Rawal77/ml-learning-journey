# Step Plot

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

### **Step Plot**

A step plot is used to show discrete changes in data. Instead of smooth transitions, it creates step-like visualizations.
It is useful for data that changes at distinct points, such as stock prices or website traffic.

```python
# Basic Step Plot Example

x = ["day1", "day2", "day3", "day4", "day5"]
y = [30, 40, 25, 30, 40]
```

plt.step(x, y)
plt.title("Basic Step Plot")
plt.xlabel("Days")
plt.ylabel("Value")
plt.show()

### **Step Plot with 'post' Argument**

'post' means the step moves forward after the change is recorded.

```python
plt.step(x, y, where="post")
plt.title("Step Plot with 'post'")
plt.xlabel("Days")
plt.ylabel("Value")
plt.show()
```

### **Step Plot with 'pre' Argument and Marker**

'pre' means the step starts before the recorded change.

Marker '.' is used to indicate the exact data points.

```python
plt.step(x, y, where="pre", marker=".")
plt.title("Step Plot with 'pre' and Markers")
plt.xlabel("Days")
plt.ylabel("Value")
plt.show()
```

### **Step Plot Using Real-World Data**

Reading an expense dataset and grouping expenses by category

```python
data = pd.read_csv('../data_sets/expense_data.csv')
group = data.groupby("Category").agg({"Amount": "sum"})
```

```python
# Creating a step plot to show total spending per category

plt.step(group.index, group["Amount"], where="mid", marker=".")
plt.title("Total Expenses by Category (Step Plot)")
plt.xlabel("Category")
plt.ylabel("Total Amount Spent")
plt.xticks(rotation=45)
plt.show()
```

### **Additional Example: Website Traffic Over a Week**

```python
days = np.arange(1, 8)  # Days of the week
visits = [150, 200, 250, 220, 270, 300, 280]

plt.step(days, visits, where="mid", marker="o", linestyle="--", color="red")
plt.title("Website Traffic Over a Week")
plt.xlabel("Days")
plt.ylabel("Number of Visits")
plt.grid(True)
plt.show()
```
