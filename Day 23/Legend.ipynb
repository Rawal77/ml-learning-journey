# **Legend in Matplotlib**

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

### Legend

Legends help in differentiating multiple data series in a plot.

They make it easier to understand the visualization by providing labels for each line, bar, or point.

```python
# Example 1: Without Legend

x = [1, 2, 3, 4, 5]
y = [45, 53, 23, 54, 22]
y1 = [34, 34, 54, 34, 43]

plt.figure(figsize=[9, 5])
plt.plot(x, y)
plt.plot(x, y1)
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.title("Line Plot Without Legend")
plt.show()
```

```python
# **With Legend**

plt.figure(figsize=[9, 5])
plt.plot(x, y, label="Female")
plt.plot(x, y1, label="Male")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.title("Line Plot With Legend")
plt.legend()  # Adds legend automatically based on labels
plt.show()
```

```python
# **Example 2: Customizing the Legend**

plt.figure(figsize=[9, 5])
plt.plot(x, y, label="Female", linestyle="--", marker="o")
plt.plot(x, y1, label="Male", linestyle="-.", marker="s")
plt.xlabel("Days")
plt.ylabel("Scores")
plt.title("Customized Legend Example")
plt.legend(loc="upper right")  # Positioning legend at the upper right
plt.show()
```

```python
# **Example 3: Adding a Shadow and Changing Font Size**

plt.figure(figsize=[9, 5])
plt.plot(x, y, label="Category A", color="blue")
plt.plot(x, y1, label="Category B", color="red")
plt.xlabel("Time")
plt.ylabel("Value")
plt.title("Legend with Shadow and Font Size")
plt.legend(fontsize=12, shadow=True)
plt.show()
```

```python
# **Example 4: Real-World Dataset - Expense Data**

data = pd.read_csv('../data_sets/expense_data.csv')
grouped = data.groupby("Category")["Amount"].sum()

plt.figure(figsize=[10, 5])
plt.bar(grouped.index, grouped.values, color=['blue', 'red', 'green', 'orange', 'purple'])
plt.xlabel("Expense Category")
plt.ylabel("Total Amount Spent")
plt.title("Expense Data with Legend")
plt.legend(["Total Expenses"], loc="upper right")  # Adding legend for clarity
plt.xticks(rotation=45)  # Rotate labels for better readability
plt.show()
```

```python
# Creating a pie chart for Marks distribution with adding legend

data = pd.read_csv("../data_sets/student.csv")
data.head()

plt.figure(figsize=(7, 7))
plt.pie(data["Marks"], labels=data["Name"], autopct="%1.1f%%", startangle=140)

# Adding a legend
plt.legend(data["Name"], title="Students", loc="upper right")

# Adding a title
plt.title("Marks Distribution Among Students")

# Show the plot
plt.show()
```
