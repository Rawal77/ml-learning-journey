# Time Series Visualization with Seaborn

Exploring trends, seasonality, and patterns using real-world datasets.

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
# Generating a synthetic time series dataset (for demonstration purposes)

dates = pd.date_range(start="2023-01-01", periods=100, freq='D')
data = pd.DataFrame({
    "Date": dates,
    "Temperature": np.random.normal(loc=20, scale=5, size=100) + np.sin(np.linspace(0, 10, 100)) * 5
})
data["Temperature"] = data["Temperature"].round(1)
```

### **1. Basic Time Series Line Plot**

```python
plt.figure(figsize=(12, 6))
sns.lineplot(x="Date", y="Temperature", data=data, marker="o", linestyle="-")
plt.title("Daily Temperature Trends")
plt.xlabel("Date")
plt.ylabel("Temperature (C)")
plt.xticks(rotation=45)
plt.show()
```

### **2. Rolling Mean (Smoothing the Time Series)**

```python
data["Rolling_Mean"] = data["Temperature"].rolling(window=7).mean()
plt.figure(figsize=(12, 6))
sns.lineplot(x="Date", y="Temperature", data=data, label="Original", alpha=0.5)
sns.lineplot(x="Date", y="Rolling_Mean", data=data, label="7-Day Rolling Mean", color="red")
plt.title("Temperature Trend with Rolling Average")
plt.xlabel("Date")
plt.ylabel("Temperature (C)")
plt.xticks(rotation=45)
plt.legend()
plt.show()
```

### **3. Handling Missing Data in Time Series**

```python

data_missing = data.copy()
data_missing.loc[np.random.choice(data_missing.index, size=10, replace=False), "Temperature"] = np.nan

plt.figure(figsize=(12, 6))
sns.lineplot(x="Date", y="Temperature", data=data_missing, marker="o", linestyle="-")
plt.title("Time Series with Missing Data")
plt.xlabel("Date")
plt.ylabel("Temperature (C)")
plt.xticks(rotation=45)
plt.show()
```

```python
# Filling missing values with rolling mean

data_missing["Temperature_Filled"] = data_missing["Temperature"].fillna(method='ffill')
plt.figure(figsize=(12, 6))
sns.lineplot(x="Date", y="Temperature_Filled", data=data_missing, marker="o", linestyle="-")
plt.title("Time Series with Missing Data (Filled)")
plt.xlabel("Date")
plt.ylabel("Temperature (C)")
plt.xticks(rotation=45)
plt.show()
```

### **4. Multiple Time Series Comparison (Stock Market Example)**

```python
dates = pd.date_range(start="2023-01-01", periods=100, freq='D')
stock_data = pd.DataFrame({
    "Date": dates,
    "Stock_A": np.cumsum(np.random.randn(100)) + 100,
    "Stock_B": np.cumsum(np.random.randn(100)) + 120
})

plt.figure(figsize=(12, 6))
sns.lineplot(x="Date", y="Stock_A", data=stock_data, label="Company A", marker="o")
sns.lineplot(x="Date", y="Stock_B", data=stock_data, label="Company B", marker="s")
plt.title("Stock Price Comparison Over Time")
plt.xlabel("Date")
plt.ylabel("Stock Price (USD)")
plt.xticks(rotation=45)
plt.legend()
plt.show()
```

### **5. Seasonal Trends & Patterns (Example: Website Traffic Analysis)**

```python
dates = pd.date_range(start="2023-01-01", periods=365, freq='D')
traffic_data = pd.DataFrame({
    "Date": dates,
    "Website_Visits": 100 + 50 * np.sin(2 * np.pi * dates.dayofyear / 365) + np.random.normal(scale=10, size=365)
})

plt.figure(figsize=(12, 6))
sns.lineplot(x="Date", y="Website_Visits", data=traffic_data)
plt.title("Website Traffic Over a Year (Seasonal Pattern)")
plt.xlabel("Date")
plt.ylabel("Number of Visits")
plt.xticks(rotation=45)
plt.show()
```
