# Data Visualization - Pie Charts with Matplotlib

```python
import matplotlib.pyplot as plt
import pandas as pd
```

###  What is a Pie Chart?

A Pie Chart is a circular statistical graphic divided into slices to illustrate numerical proportions.

It is useful for showing percentage distributions of a dataset.

### Example 1: Basic Pie Chart

Scenario: Showing the popularity of different noodle brands in a survey.

```python
# Data
noodles = ['Wai Wai', 'Rum Pum', 'MaMa', 'Current', '2pm']
x = [35, 12, 41, 37, 30]  # Percentage distribution of survey results
colors = ['red', 'green', 'yellow', 'orange', 'blue']

# Creating the Pie Chart
plt.figure(figsize=(6,6))
plt.pie(x, labels=noodles, colors=colors)

# Title
plt.title("Popularity of Noodle Brands (Survey Data)")

# Display the plot
plt.show()
```

### Example 2: Exploding a Slice

Scenario: Highlighting the most popular noodle brand in the chart.

```python
explode_values = [0, 0, 0, 0, 0.1]  # Slightly pull out the last slice (2pm)

plt.figure(figsize=(6,6))
plt.pie(x, labels=noodles, colors=colors, explode=explode_values)

# Title
plt.title("Popularity of Noodle Brands (Highlighted)")

plt.show()
```

### Example 3: Adding Percentage Labels (autopct)

Scenario: Displaying exact percentage values on the chart.

```python
plt.figure(figsize=(6,6))
plt.pie(x, labels=noodles, colors=colors, explode=explode_values, autopct="%.2f%%")  # Formatting to 2 decimal places

# Title
plt.title("Noodle Brand Popularity (With Percentage)")

plt.show()
```

### Example 4: Adjusting the Start Angle

Scenario: Rotating the Pie Chart for better readability.

```python
plt.figure(figsize=(6,6))
plt.pie(x, labels=noodles, colors=colors, explode=explode_values, autopct="%.2f%%", startangle=180)

# Title
plt.title("Noodle Brand Popularity (Rotated)")

plt.show()
```

### Example 5: Real-World Dataset - Payment Methods

Scenario: Visualizing the distribution of different payment methods from a dataset.

```python
# Load dataset (assumed available)
df = pd.read_csv('../data_sets/expense_data.csv')

# Grouping data by Payment Mode and summing the amounts
grouped_by_payment = df.groupby("Payment Mode")["Amount"].sum()

# Creating a Pie Chart
plt.figure(figsize=(7,7))
plt.pie(grouped_by_payment.values, labels=grouped_by_payment.index, autopct="%.1f%%", startangle=90, colors=['cyan', 'pink', 'purple', 'grey'])

# Title
plt.title("Distribution of Expenses by Payment Mode")

plt.show()
```

### Example 6: Another Real-World Example - Sales by Product Category

Scenario: Showing sales distribution across different product categories.

```python
# Grouping by category
grouped_by_category = df.groupby("Category")["Amount"].sum()

# Creating a Pie Chart
plt.figure(figsize=(7,7))
plt.pie(grouped_by_category.values, labels=grouped_by_category.index, autopct="%.1f%%", startangle=140, colors=['red', 'blue', 'green', 'yellow'])

# Title
plt.title("Sales Distribution by Product Category")

plt.show()
```

### Example 7: Market Share of Smartphone Brands

Scenario: A research firm conducted a study to analyze the global smartphone market share of major brands.

```python
# Defining smartphone brands and their global market share percentage

brands = ['Apple', 'Samsung', 'Xiaomi', 'Oppo', 'Vivo', 'Others']
market_share = [30, 28, 15, 10, 9, 8]  # Market share in percentage
colors = ['gold', 'blue', 'red', 'green', 'purple', 'grey']  # Assigning unique colors
explode = [0.1, 0, 0, 0, 0, 0]  # Highlighting Apple for better visualization

# Creating the Pie Chart
plt.figure(figsize=(8,8))
plt.pie(market_share, labels=brands, autopct="%.1f%%", startangle=140, explode=explode, colors=colors)

# Adding a title
plt.title("Global Smartphone Market Share (2024)")

# Display the chart
plt.show()
```
