# Data Visualization with Matplotlib â€“ Subplots

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

### Subplots

Subplots allow you to display multiple plots in a single figure.

They are essential when you need to compare different visualizations side by side,

saving space and making it easier to spot trends and relationships across datasets.

```python
x = [1, 2, 3, 4, 5]
y_age = [45, 43, 54, 23, 43]
y_weight = [43, 32, 43, 54, 43]

plt.figure(figsize=[10, 5])
```

```python
# First subplot: Age

plt.subplot(1, 2, 1)
plt.plot(x, y_age, marker='o', color='blue')
plt.title("Age")
plt.xlabel("X-axis")
plt.ylabel("Age Values")
```

```python
# Second subplot: Weight

plt.subplot(1, 2, 2)
plt.plot(x, y_weight, marker='s', color='green')
plt.title("Weight")
plt.xlabel("X-axis")
plt.ylabel("Weight Values")

plt.tight_layout()
plt.show()
```

### Example 2: Creating a 2x2 Grid of Different Plot Types

```python
plt.figure(figsize=[12, 10])

# Top-left: Line Plot (Sine wave)
x = np.linspace(0, 10, 100)
y = np.sin(x)
plt.subplot(2, 2, 1)
plt.plot(x, y, label="Sine", color="red")
plt.title("Line Plot: Sine Wave")
plt.xlabel("X-axis")
plt.ylabel("Amplitude")
plt.legend()

# Top-right: Histogram

data = np.random.randn(1000)
plt.subplot(2, 2, 2)
plt.hist(data, bins=30, color="purple", edgecolor="black", alpha=0.7)
plt.title("Histogram")
plt.xlabel("Value")
plt.ylabel("Frequency")

# Bottom-left: Scatter Plot

x_scatter = np.random.rand(50)
y_scatter = np.random.rand(50)
plt.subplot(2, 2, 3)
plt.scatter(x_scatter, y_scatter, color="orange")
plt.title("Scatter Plot")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")

# Bottom-right: Bar Chart

categories = ['A', 'B', 'C', 'D']
values = [20, 35, 30, 35]
plt.subplot(2, 2, 4)
plt.bar(categories, values, color="cyan")
plt.title("Bar Chart")
plt.xlabel("Category")
plt.ylabel("Value")

plt.tight_layout()
plt.show()
```

### Example 3: Subplots with a Real-World Dataset (Expense Data)

```python
# Assume the dataset 'expense_data.csv' has columns: 'Month', 'Amount', 'Payment Mode'

data = pd.read_csv('../data_sets/expense_data.csv')
# Display first few rows
print(data.head())
```

```python
# Creating two subplots:
# Subplot 1: Line Plot of expense Amount over Months
# Subplot 2: Bar Chart of total Amount per Payment Mode

plt.figure(figsize=[12, 5])

# Subplot 1: Line Plot
plt.subplot(1, 2, 1)
plt.plot(data['Date'], data['Amount'], marker='o', color='magenta')
plt.title("Daily Expense Trend")
plt.xlabel("Daily")
plt.ylabel("Expense Amount (USD)")
plt.xticks(rotation=45)

# Subplot 2: Bar Chart
grouped_expense = data.groupby("Payment Mode")["Amount"].sum()
plt.subplot(1, 2, 2)
plt.bar(grouped_expense.index, grouped_expense.values, color=['blue', 'green', 'orange', 'red'])
plt.title("Total Expenses by Payment Mode")
plt.xlabel("Payment Mode")
plt.ylabel("Total Amount (USD)")

plt.tight_layout()
plt.show()
```

### Example 4: Subplots with a Real-World Dataset (Student Data)

```python
# Creating a sample student dataset (from provided values)
students_data = {
    "Name": ["Rama", "Ram", "Shyam", "Hari"],
    "Marks": [98, 39, 77, 68],
    "Semester": ["VII", "V", "VII", "IX"],
    "Grade": ["A", "D", "B", "B"],
    "Place": ["Kathmandu", "Bhaktapur", "Lalitpur", "Nuwakot"]
}
df_students = pd.DataFrame(students_data)

plt.figure(figsize=[12, 6])
# Subplot 1: Bar Chart of Marks by Student
plt.subplot(1, 2, 1)
plt.bar(df_students["Name"], df_students["Marks"], color="skyblue")
plt.title("Student Marks")
plt.xlabel("Student Name")
plt.ylabel("Marks")

# Subplot 2: Pie Chart of Grade Distribution
# Calculate counts per grade
grade_counts = df_students["Grade"].value_counts()
plt.subplot(1, 2, 2)
plt.pie(grade_counts.values, labels=grade_counts.index, autopct="%.1f%%", startangle=90)
plt.title("Grade Distribution")

plt.tight_layout()
plt.show()
```
